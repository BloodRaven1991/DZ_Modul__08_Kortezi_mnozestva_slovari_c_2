Задание 1
Создайте программу, хранящую информацию о великих баскетболистах. Нужно хранить ФИО баскетболиста и
его рост. Требуется реализовать возможность добавления,
удаления, поиска, замены данных. Используйте словарь
для хранения информации.


```python
basket_players = {}

def add_player(name, height):
    basket_players[name] = height
    print(f"Добавлен баскетболист: {name}, рост: {height} см.")

def remove_p(name):
    if name in basket_players:
        del basket_players[name]
        print(f"Удален баскетболист: {name}.")
    else:
        print(f"Баскетболист с именем {name} не найден.")

def search_p(name):
    if name in basket_players:
        print(f"Поиск баскетболиста: {name}, рост: {basket_players[name]} см.")
    else:
        print(f"Баскетболист с именем {name} не найден.")

def update_p(name, height):
    if name in basket_players:
        basket_players[name] = height
        print(f"Данные обновлены для {name}. Новый рост: {height} см.")
    else:
        print(f"Баскетболист с именем {name} не найден.")

def display_p():
    if basket_players:
        print("Список баскетболистов:")
        for name, height in basket_players.items():
            print(f"{name}: {height} см")
    else:
        print("Список баскетболистов пуст.")


while True:
    print("\nМеню:")
    print("1. Добавить")
    print("2. Удалить")
    print("3. Поиск")
    print("4. Обновить")
    print("5. Показать всех")
    print("6. Выход")

    choice = input("Выберите действие (1-6): ")

    if choice == '1':
        name = input("Введите ФИО баскетболиста: ")
        height = int(input("Введите рост (в см): "))
        add_player(name, height)
    elif choice == '2':
        name = input("Введите ФИО баскетболиста для удаления: ")
        remove_p(name)
    elif choice == '3':
        name = input("Введите ФИО баскетболиста для поиска: ")
        search_p(name)
    elif choice == '4':
        name = input("Введите ФИО баскетболиста для обновления данных: ")
        height = int(input("Введите новый рост (в см): "))
        update_p(name, height)
    elif choice == '5':
        display_p()
    elif choice == '6':
        print("Выход из программы.")
        break
    else:
        print("Неверный ввод. Пожалуйста, выберите число от 1 до 6.")
```

Задание 2
Создайте программу «Англо-французский словарь».
Нужно хранить слово на английском языке и его перевод
на французский. Требуется реализовать возможность добавления, удаления, поиска, замены данных. Используйте
словарь для хранения информации.

```python
class EnglishFrenchDictionary:
    def __init__(self):
        self.dictionary = {}

    def add_word(self, english, french):
        """Добавляет новое слово в словарь."""
        if english in self.dictionary:
            print("Слово уже существует. Обновите перевод вместо добавления.")
        else:
            self.dictionary[english] = french
            print(f"Слово '{english}' добавлено с переводом '{french}'.")

    def remove_word(self, english):
        """Удаляет слово из словаря."""
        if english in self.dictionary:
            del self.dictionary[english]
            print(f"Слово '{english}' удалено.")
        else:
            print("Слово не найдено.")

    def search_word(self, english):
        """Ищет перевод слова в словаре."""
        if english in self.dictionary:
            print(f"{english} -> {self.dictionary[english]}")
        else:
            print("Слово не найдено.")

    def update_word(self, english, new_french):
        """Обновляет перевод слова в словаре."""
        if english in self.dictionary:
            self.dictionary[english] = new_french
            print(f"Перевод слова '{english}' обновлен на '{new_french}'.")
        else:
            print("Слово не найдено.")

    def display_dictionary(self):
        """Отображает все слова и их переводы."""
        if not self.dictionary:
            print("Словарь пуст.")
        else:
            for english, french in self.dictionary.items():
                print(f"{english} -> {french}")

def main():
    dictionary = EnglishFrenchDictionary()

    while True:
        print("\nМеню:")
        print("1. Добавить слово")
        print("2. Удалить слово")
        print("3. Найти слово")
        print("4. Обновить перевод слова")
        print("5. Показать словарь")
        print("6. Выход")

        choice = input("Выберите действие (1-6): ")

        if choice == '1':
            english = input("Введите слово на английском: ")
            french = input("Введите перевод на французском: ")
            dictionary.add_word(english, french)
        elif choice == '2':
            english = input("Введите слово для удаления: ")
            dictionary.remove_word(english)
        elif choice == '3':
            english = input("Введите слово для поиска: ")
            dictionary.search_word(english)
        elif choice == '4':
            english = input("Введите слово для обновления: ")
            new_french = input("Введите новый перевод: ")
            dictionary.update_word(english, new_french)
        elif choice == '5':
            dictionary.display_dictionary()
        elif choice == '6':
            print("Выход из программы.")
            break
        else:
            print("Некорректный выбор. Пожалуйста, попробуйте снова.")

if __name__ == "__main__":
    main()
```

Задание 3
Создайте программу «Фирма». Нужно хранить информацию о человеке: ФИО, телефон, рабочий email,
название должности, номер кабинета, skype. Требуется
реализовать возможность добавления, удаления, поиска, замены данных. Используйте словарь для хранения
информации.

```python
class Employee:
    def __init__(self):
        self.employees = {}

    def add_employee(self, name, phone, email, position, office, skype):
        """Добавляет нового сотрудника."""
        if name in self.employees:
            print("Сотрудник с таким именем уже существует. Обновите данные вместо добавления.")
        else:
            self.employees[name] = {
                'phone': phone,
                'email': email,
                'position': position,
                'office': office,
                'skype': skype
            }
            print(f"Сотрудник '{name}' добавлен.")

    def remove_employee(self, name):
        """Удаляет сотрудника."""
        if name in self.employees:
            del self.employees[name]
            print(f"Сотрудник '{name}' удален.")
        else:
            print("Сотрудник не найден.")

    def search_employee(self, name):
        """Ищет информацию о сотруднике."""
        if name in self.employees:
            info = self.employees[name]
            print(f"ФИО: {name}, Телефон: {info['phone']}, Email: {info['email']}, "
                  f"Должность: {info['position']}, Кабинет: {info['office']}, Skype: {info['skype']}")
        else:
            print("Сотрудник не найден.")

    def update_employee(self, name, phone=None, email=None, position=None, office=None, skype=None):
        """Обновляет данные о сотруднике."""
        if name in self.employees:
            if phone:
                self.employees[name]['phone'] = phone
            if email:
                self.employees[name]['email'] = email
            if position:
                self.employees[name]['position'] = position
            if office:
                self.employees[name]['office'] = office
            if skype:
                self.employees[name]['skype'] = skype
            print(f"Данные сотрудника '{name}' обновлены.")
        else:
            print("Сотрудник не найден.")

    def display_employees(self):
        """Отображает всех сотрудников."""
        if not self.employees:
            print("Список сотрудников пуст.")
        else:
            for name, info in self.employees.items():
                print(f"ФИО: {name}, Телефон: {info['phone']}, Email: {info['email']}, "
                      f"Должность: {info['position']}, Кабинет: {info['office']}, Skype: {info['skype']}")


def main():
    company = Employee()

    while True:
        print("\nМеню:")
        print("1. Добавить сотрудника")
        print("2. Удалить сотрудника")
        print("3. Найти сотрудника")
        print("4. Обновить данные о сотруднике")
        print("5. Показать всех сотрудников")
        print("6. Выход")

        choice = input("Выберите действие (1-6): ")

        if choice == '1':
            name = input("Введите ФИО сотрудника: ")
            phone = input("Введите телефон: ")
            email = input("Введите рабочий email: ")
            position = input("Введите должность: ")
            office = input("Введите номер кабинета: ")
            skype = input("Введите Skype: ")
            company.add_employee(name, phone, email, position, office, skype)
        
        elif choice == '2':
            name = input("Введите ФИО сотрудника для удаления: ")
            company.remove_employee(name)
        
        elif choice == '3':
            name = input("Введите ФИО сотрудника для поиска: ")
            company.search_employee(name)
        
        elif choice == '4':
            name = input("Введите ФИО сотрудника для обновления: ")
            phone = input("Введите новый телефон (или оставьте пустым для игнорирования): ")
            email = input("Введите новый рабочий email (или оставьте пустым для игнорирования): ")
            position = input("Введите новую должность (или оставьте пустым для игнорирования): ")
            office = input("Введите новый номер кабинета (или оставьте пустым для игнорирования): ")
            skype = input("Введите новый Skype (или оставьте пустым для игнорирования): ")
            company.update_employee(name, phone if phone else None,
                                   email if email else None,
                                   position if position else None,
                                   office if office else None,
                                   skype if skype else None)
        
        elif choice == '5':
            company.display_employees()
        
        elif choice == '6':
            print("Выход из программы.")
            break
        
        else:
            print("Некорректный выбор. Пожалуйста, попробуйте снова.")


if __name__ == "__main__":
    main()
```

Задание 4
Создайте программу «Книжная коллекция». Нужно
хранить информацию о книгах: автор, название книги,
жанр, год выпуска, количество страниц, издательство.
Требуется реализовать возможность добавления, удаления, поиска, замены данных. Используйте словарь для
хранения информации.

```python
class BookCollection:
    def __init__(self):
        self.books = {}

    def add_book(self, title, author, genre, year, pages, publisher):
        """Добавляет новую книгу в коллекцию."""
        if title in self.books:
            print("Книга с таким названием уже существует. Обновите данные вместо добавления.")
        else:
            self.books[title] = {
                'author': author,
                'genre': genre,
                'year': year,
                'pages': pages,
                'publisher': publisher
            }
            print(f"Книга '{title}' добавлена.")

    def remove_book(self, title):
        """Удаляет книгу из коллекции."""
        if title in self.books:
            del self.books[title]
            print(f"Книга '{title}' удалена.")
        else:
            print("Книга не найдена.")

    def search_book(self, title):
        """Ищет информацию о книге."""
        if title in self.books:
            info = self.books[title]
            print(f"Название: {title}, Автор: {info['author']}, Жанр: {info['genre']}, "
                  f"Год выпуска: {info['year']}, Количество страниц: {info['pages']}, Издательство: {info['publisher']}")
        else:
            print("Книга не найдена.")

    def update_book(self, title, author=None, genre=None, year=None, pages=None, publisher=None):
        """Обновляет данные о книге."""
        if title in self.books:
            if author:
                self.books[title]['author'] = author
            if genre:
                self.books[title]['genre'] = genre
            if year:
                self.books[title]['year'] = year
            if pages:
                self.books[title]['pages'] = pages
            if publisher:
                self.books[title]['publisher'] = publisher
            print(f"Данные о книге '{title}' обновлены.")
        else:
            print("Книга не найдена.")

    def display_books(self):
        """Отображает все книги в коллекции."""
        if not self.books:
            print("Книжная коллекция пуста.")
        else:
            for title, info in self.books.items():
                print(f"Название: {title}, Автор: {info['author']}, Жанр: {info['genre']}, "
                      f"Год выпуска: {info['year']}, Количество страниц: {info['pages']}, "
                      f"Издательство: {info['publisher']}")

def main():
    collection = BookCollection()

    while True:
        print("\nМеню:")
        print("1. Добавить книгу")
        print("2. Удалить книгу")
        print("3. Найти книгу")
        print("4. Обновить данные о книге")
        print("5. Показать все книги")
        print("6. Выход")

        choice = input("Выберите действие (1-6): ")

        if choice == '1':
            title = input("Введите название книги: ")
            author = input("Введите имя автора: ")
            genre = input("Введите жанр: ")
            year = input("Введите год выпуска: ")
            pages = input("Введите количество страниц: ")
            publisher = input("Введите издательство: ")
            collection.add_book(title, author, genre, year, pages, publisher)

        elif choice == '2':
            title = input("Введите название книги для удаления: ")
            collection.remove_book(title)

        elif choice == '3':
            title = input("Введите название книги для поиска: ")
            collection.search_book(title)

        elif choice == '4':
            title = input("Введите название книги для обновления: ")
            author = input("Введите нового автора (или оставьте пустым для игнорирования): ")
            genre = input("Введите новый жанр (или оставьте пустым для игнорирования): ")
            year = input("Введите новый год выпуска (или оставьте пустым для игнорирования): ")
            pages = input("Введите новое количество страниц (или оставьте пустым для игнорирования): ")
            publisher = input("Введите новое издательство (или оставьте пустым для игнорирования): ")
            collection.update_book(title, author if author else None,
                                   genre if genre else None,
                                   year if year else None,
                                   pages if pages else None,
                                   publisher if publisher else None)

        elif choice == '5':
            collection.display_books()

        elif choice == '6':
            print("Выход из программы.")
            break

        else:
            print("Некорректный выбор. Пожалуйста, попробуйте снова.")

if __name__ == "__main__":
    main()
```
